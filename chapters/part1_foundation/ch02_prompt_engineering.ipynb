## 예제 2-1. 역할·맥락·지시·형식 프롬프트 빌더

```python
def build_prompt(role, context, instruction, output_format):
    return (
        f"[역할] {role}\n"
        f"[맥락] {context}\n"
        f"[지시] {instruction}\n"
        f"[출력 형식] {output_format}"
    )

prompt = build_prompt(
    role="결제 시스템을 담당하는 백엔드 개발자",
    context="결제 API에서 500 에러가 간헐적으로 발생한다",
    instruction="원인을 가능성이 높은 순서로 세 가지 제시하라",
    output_format="원인, 근거, 확인 방법 세 열을 가진 표",
)
print(prompt)
```

## 예제 2-2. Few-shot 분류 프롬프트 빌더

```python
examples = [
    ("환불받고 싶어요", "환불"),
    ("아직도 안 왔어요", "배송"),
    ("색상을 바꿀 수 있나요", "기타"),
]

def build_fewshot(examples, query):
    shots = "\n".join(f"문의: {x}\n분류: {y}" for x, y in examples)
    return f"{shots}\n문의: {query}\n분류:"

print(build_fewshot(examples, "돈을 다시 돌려주세요"))
```

## 예제 2-3. 군말을 걸러 내는 안전한 JSON 파서

```python
import json

def parse_json(text):
    try:
        return json.loads(text)
    except json.JSONDecodeError:
        start = text.find("{")
        end = text.rfind("}")
        if start != -1 and end != -1:
            return json.loads(text[start:end + 1])
        raise

raw = '결과는 다음과 같습니다: {"분류": "환불", "확신도": 0.92}'
print(parse_json(raw))
```
