## 예제 1-1. softmax와 온도에 따른 확률 분포

```python
import numpy as np

# 모델이 "오늘 하늘이 정말" 다음에 매긴 점수(로짓)를 가정한다
words = ["맑다", "파랗다", "높다", "슬프다", "계산기"]
logits = np.array([3.2, 2.8, 1.5, -0.5, -2.0])

def softmax(logits, temperature):
    z = logits / temperature
    z = z - z.max()              # 수치 안정화
    exp = np.exp(z)
    return exp / exp.sum()

for temperature in [0.2, 1.0, 2.0]:
    probs = softmax(logits, temperature)
    line = ", ".join(f"{w} {p:.2f}" for w, p in zip(words, probs))
    print(f"온도 {temperature}: {line}")
```

## 예제 1-2. tiktoken으로 한국어·영어 토큰 수 비교

```python
# 코랩에서 먼저: pip install tiktoken
import tiktoken

enc = tiktoken.get_encoding("cl100k_base")

samples = {
    "영어": "The capital of Korea is Seoul.",
    "한국어": "대한민국의 수도는 서울입니다.",
}
for label, text in samples.items():
    n_token = len(enc.encode(text))
    print(f"{label}: {n_token}개 토큰  ({len(text)}글자)")
```
