## 예제 10-1. 조사·작성·검수 파이프라인

```python
def agent(role, task):
    return model.generate_content(f"너는 {role}이다.\n{task}").text

topic = "사내 점심시간 안내문"
research = agent("사실을 정리하는 조사원", f"{topic}에 필요한 핵심 사실을 한 줄로 정리하라")
draft = agent("간결한 작성자", f"다음 사실로 안내문 초안을 한 문장으로 써라.\n{research}")
review = agent("꼼꼼한 검수자", f"다음 초안에서 고칠 점을 한 줄로 지적하라.\n{draft}")

print("[조사]", research)
print("[작성]", draft)
print("[검수]", review)
```

## 예제 10-2. 휴먼 인 더 루프 승인 게이트

```python
def send_notice(text):
    return f"발송 완료: {text}"

def with_approval(action, payload):
    print("[확인 요청]", payload)
    ok = input("실행할까요? (y/n) ")        # 코랩에서 사람이 입력
    if ok.strip().lower() == "y":
        return action(payload)
    return "사람이 취소함"

print(with_approval(send_notice, draft))
```
