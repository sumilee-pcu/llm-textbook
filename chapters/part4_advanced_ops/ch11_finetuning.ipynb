## 예제 11-1. 파인튜닝 학습 데이터 만들기

```python
import json

examples = [
    {"input": "환불 문의가 왔어", "output": "고객님, 환불은 구매 후 7일 이내 가능합니다."},
    {"input": "배송 언제 와", "output": "고객님, 배송은 보통 2~3일 걸립니다."},
    {"input": "교환하고 싶어", "output": "고객님, 교환은 미사용 제품에 한해 가능합니다."},
]

with open("train.jsonl", "w", encoding="utf-8") as f:
    for ex in examples:
        f.write(json.dumps(ex, ensure_ascii=False) + "\n")

print("학습 예시:", len(examples), "개")
print("형식 예:", json.dumps(examples[0], ensure_ascii=False))
```

## 예제 11-2. 프롬프트·RAG·파인튜닝 선택

```python
def choose_method(needs_knowledge, needs_style, budget_low):
    if budget_low or not (needs_knowledge or needs_style):
        return "프롬프트"        # 가장 싸고 빠르게 먼저
    if needs_knowledge:
        return "RAG"             # 모델에 없는 지식은 검색으로
    return "파인튜닝"            # 말투·형식 일관성은 학습으로

print(choose_method(needs_knowledge=True,  needs_style=False, budget_low=False))
print(choose_method(needs_knowledge=False, needs_style=True,  budget_low=False))
print(choose_method(needs_knowledge=False, needs_style=False, budget_low=True))
```
